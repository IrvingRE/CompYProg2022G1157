#Problema 3.1

Escriba un diagrama de flujo tal que dado como datos 270 números enteros, ob­tenga la suma de los números impares y el promedio de los números pares.

````python
SUMPAR=0
SUMIMP=0
CUEPAR=0
i=1

while i<=270:
    NUM=i
    if NUM>0:
        if NUM%2!=0:
            SUMPAR=SUMPAR+NUM
            CUEPAR=CUEPAR+1
        else:
            SUMIMP=SUMIMP+NUM
    i=i+1
PROPAR=SUMPAR / CUEPAR
print(f"La suma de los números impares es: {SUMIMP} , El promedio de los números pares es: {PROPAR} ")
````

#Problema 3.4
Calcule el aumento de sueldo para un grupo de empleados de una empresa te­niendo en cuenta el siguiente criterio:
```
Si el sueldo es inferior a $1000  : Aumento 15%
Si el sueldo es mayor o igual a $1000  : Aumento 12%
```
Imprima el sueldo nuevo del trabajador y el total de nómina de la empresa, 
considerando este nuevo aumento.


````python
NOM=0
SUE=float(input("¿De cuanto es el sueldo del empleado?  "))
while SUE!=-1:
    if SUE<1000:
        NSUE=round(SUE*1.15,2)
    else:
        NSUE=round(SUE*1.12,2)
    NOM=round(NOM+NSUE,2)
    print(f"El sueldo nuevo del trabajador es de: {NSUE}")
    SUE=float(input("¿De cuanto es el sueldo del empleado?  "))
print(f"El total de nómina de la empresa es: {NOM}")
````
